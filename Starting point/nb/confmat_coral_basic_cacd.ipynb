```python
import pandas as pd
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# CACD prediction analysis

```python
with open('../model-code/cacd_coral_basic/test_predictions.log','r') as filehandle:
    predictions = np.array([int(i) for i in filehandle.read().split(',')])
groundtruth = np.array(list(pd.read_csv('../datasets/cacd_test.csv')['age']))
```

```python
# figure, axs = plt.subplot()

hist = sns.histplot(predictions,kde=True,bins=50)
plt.xticks(rotation=90)
plt.title('CACD predictions with basic coral distribution')
plt.xlabel('Age')
plt.show()
```

```python
hist = sns.histplot(groundtruth,kde=True,bins=50)
plt.xticks(rotation=90)
plt.title('CACD groundtruth distribution')
plt.xlabel('Age')
plt.show()
```

```python
cm = confusion_matrix(groundtruth, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=list(set(groundtruth)))
disp.plot(include_values=False)
plt.show()
```
